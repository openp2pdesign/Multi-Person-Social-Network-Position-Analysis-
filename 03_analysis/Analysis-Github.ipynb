```python
first_person_color = '#fa824c'
second_person_color = '#00bdbf'
third_person_color = '#f3f1e9'
```

```python
author = 'openp2pdesign'
```

## Following / Friends

```python
# Load graph
graph = nx.read_graphml("data/github.graphml")
```

```python
binsn = 20
```

```python
df = pd.DataFrame.from_dict(dict(graph.nodes(data=True)), orient='index')
```

```python
df.columns
```

## Betweenness

```python
# Plot the data
what = 'Betweenness Centrality'
ax = df[what].plot(kind="hist", figsize=(18,5), bins=binsn, color=second_person_color, edgecolor='black', legend=False)

ax.axvline(df.loc[author][what], lw=4, color=first_person_color)

plt.ylabel('Frequency', fontsize=16)
plt.xlabel(what, fontsize=16)
plt.savefig("png/"+what.lower().replace(" ", "_")+"-GITHUB.png", format='png', dpi=300, bbox_inches="tight")
plt.savefig("svg/"+what.lower().replace(" ", "_")+"-GITHUB.svg", bbox_inches="tight")
```

```python
author_community = ''
for row in df.iterrows():
    if row[1]["label"] == author:
        try:
            dfc.loc[row[1]['label']] = ['first-person', new_col[row[1]["Modularity Class"]]]
        except:
            dfc.loc[row[1]['label']] = ['first-person', row[1]["Modularity Class"]]
    else:
        try:
            dfc.loc[row[1]['label']] = ['second-person', new_col[row[1]["Modularity Class"]]]
        except:
            dfc.loc[row[1]['label']] = ['second-person', row[1]["Modularity Class"]]
    #dfc = pd.concat([new_row,dfc.loc[:]]).reset_index(drop=True)
```

```python
what = 'Largest 10 Communities of the Author + Team'
ax = dfc[dfc['Modularity Class'].isin(cols)]["Modularity Class"].value_counts().plot(kind="bar", figsize=(30,10), color=gephi_palette)
ax.bar_label(ax.containers[0])
plt.xlabel('Communities')
plt.ylabel('Twitter accounts')

for i in df.iterrows():
    if i[1]["label"] == author:
        x_labels = list(ax.get_xticklabels())
        for i in x_labels:
            if i.get_text() == author_community:
                ax.axvline(i.get_position()[0], color=first_person_color, lw=4, zorder=1) 

plt.savefig("png/"+what.lower().replace(" ", "_")+"-GITHUB.png", format='png', dpi=300, bbox_inches="tight")
plt.savefig("svg/"+what.lower().replace(" ", "_")+"-GITHUB.svg", bbox_inches="tight")
```
