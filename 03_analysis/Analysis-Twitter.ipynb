## Following / Friends

```python
# Load graph
graph = nx.read_graphml("data/twitter.graphml")
```

```python
binsn = 20
```

```python
df = pd.DataFrame.from_dict(dict(graph.nodes(data=True)), orient='index')
```

```python
df.columns
```

## Betweenness

```python
# Plot the data
what = 'Betweenness Centrality'
ax = df[what].plot(kind="hist", figsize=(18,5), bins=binsn, color=third_person_color, edgecolor='black', legend=False)

for index, row in df.iterrows():
    if row.perspective == "first-person":
        ax.axvline(row[what], lw=4, color=first_person_color)
    elif row.perspective == "second-person":
        ax.axvline(row[what], lw=1, color=second_person_color, zorder=1)

plt.ylabel('Frequency', fontsize=16)
plt.xlabel(what, fontsize=16)
plt.savefig("png/"+what.lower().replace(" ", "_")+"-TWITTER.png", format='png', dpi=300, bbox_inches="tight")
plt.savefig("svg/"+what.lower().replace(" ", "_")+"-TWITTER.svg", bbox_inches="tight")
```
